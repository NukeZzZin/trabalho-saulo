## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))
    
    if valor == 0:
        break # Sai do loop quando o usuário digitar zero
        
    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 1: Somatória dos Pares
Peça um número `N` ao usuário. Use um loop `for` ou `while` para calcular e exibir a SOMA de todos os números PARES entre 1 e `N`.

```python
N = int(input("Digite o valor final N: "))
# Seu código aqui
resultado = 0

for index in range(1, N + 1):
    if index % 2 == 0:
        resultado += index

print("Soma dos pares:", resultado)

```

### Desafio 26. 
Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
# Escreva seu código aqui
eleitores = int(input("Total de eleitores: "))
acumulador_1 = acumulador_2 = acumulador_3 = 0

for _index in range(eleitores):
    entrada = int(input("Vote (1,2,3): "))
    if entrada == 1:
        acumulador_1 += 1
    elif entrada == 2:
        acumulador_2 += 1
    else:
        acumulador_3 += 1

print("Candidato 1:", acumulador_1)
print("Candidato 2:", acumulador_2)
print("Candidato 3:", acumulador_3)
```

### Desafio 27. 
Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade de turmas e a quantidade de alunos para cada turma. As turmas não podem ter mais de 40 alunos.

```python
# Escreva seu código aqui
turmas = int(input("Número de turmas: "))
total = 0

for _ in range(turmas):
    alunos = int(input("Alunos (máx 40): "))
    while alunos > 40:
        alunos = int(input("Inválido, digite novamente: "))
    total += alunos

print("Média de alunos:", total / turmas)
```
